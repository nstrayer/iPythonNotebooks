##How can this data be visualized geographically?

* Kindly Lewis Mitchell has provided us with the raw data on the happiness of hundreds of cities around the country. We will utilize this data to visualize the distribution of happiness and compare it with the distribution of other factors.  We are looking to piece together geographic correlations of factors with happiness.

## First we load in the happiness data provided to us and generate some prelimiary plots: 

```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt 
import numpy, scipy.stats
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import rgb2hex
from matplotlib.patches import Polygon
niceblu = "#424FA4" #Stolen from Prof. Bagrow's notes. Thanks!

df = pd.read_csv("happiestCities2011.csv")
df["tweetNumber"] = list(df["tweetNumber"])[::-1] #The data was accidentally sent to us reversed, this fixes that. 


df.head()
```

Now that the data is loaded in let's orient ourselves with the data with a nice scatter plot of the number of tweets to happiness:

```python
plt.plot(df["happiness"],df["tweetNumber"],'o', c = niceblu)

highVal = df[(df['tweetNumber'] > 450000)]

for entry in highVal.index.values: 
    plt.annotate(highVal.city[entry], xy=(highVal.happiness[entry], highVal.tweetNumber[entry]), \
                 xytext=((highVal.happiness[entry] + .005, highVal.tweetNumber[entry]-1000)))

plt.annotate(df.city[0], xy=(df.happiness[0], df.tweetNumber[0]), \
             xytext=((df.happiness[0]-.01, df.tweetNumber[0]+40000)))

plt.annotate(df.city[len(df)-1], xy=(df.happiness[len(df)-1], df.tweetNumber[len(df)-1]), \
             xytext=((df.happiness[len(df)-1] - .017, df.tweetNumber[len(df)-1]+40000)))

plt.title("Number of tweets to happiness plot",fontsize = 15)
plt.ylabel("Number of tweets", fontsize = 15)
plt.xlabel("Happiness value",fontsize = 15)
plt.xlim([5.8,6.27])
plt.gcf().set_size_inches(16,10)
```

### Take from this plot: 

* As is to be expected large metro areas dominate the most tweets. 

* Napa, CA is the happiest city. 

* Beaumont, TX is the saddest. These match the paper published from this data so that is good. 

* An interesting bell curve shape.

## Now let's plot this on a map: 

* We need a way to take the data from simply city names to latitude and longitude coordinates for plotting in basemap: 

* To do this we registered with Google and used their geolocations API. Feel free to take this function and use your own API key! 

```python
import requests
import json

def make_api_string(term): 
    """Simplifies string to work with API request format.
    """
    res = term.replace(", ","+")
    return res

def geolocations(city):
    """Give me a city string and I will give you the lat and lon coordinates of it.
    """
    API_KEY = "AIzaSyByolaM8l2clFXLvBhaIRE3daYbFwK8l4Y" #our very own API key!

    #The standard JSON call function doesnt work so I made an ugly one, Luckily were only doing one thing with the api.
    url_call = "https://maps.googleapis.com/maps/api/geocode/json?address=%s&sensor=false&key=%s" % (make_api_string(city), API_KEY)
    r = requests.get(url_call) 
    results = json.loads(r.text)
    
    lat = results['results'][0]['geometry']['location']['lat'] #Why oh why so many dictionaries?!
    lon = results['results'][0]['geometry']['location']['lng']
    return (lat,lon)

geolocations("Burlington, VT") 
```

## Now that we have a way to get geographic data let's merge the datasets and plot!

```python
try: 
    df = pd.read_csv("happiestCitiesLatLon.csv") #Cacheing is the solution to the world's problems. 
    del df['Unnamed: 0'] #funny that pandas doesnt read its own csvs well. 
except: 
    lats = []
    lons = []
    for entry in df['city']: 
        lats.append(geolocations(entry)[0]) 
        lons.append(geolocations(entry)[1]) 
    
    df['lat'] = lats
    df['lon'] = lons
    df.to_csv('happiestCitiesLatLon.csv')
        
df.head() #test to see if it worked.
```

```python
# Create a dictionary called 'stateFullName' which contains the full name of each state along
# with the happiness average.

df = pd.read_csv("happiestCitiesLatLon.csv")

states = {}  #get all the state values into a dictionary keyed by the state name. 
for i in range(len(df)):
    state = df['city'][i][-2:] 
    try: 
        states[state].append(df['happiness'][i])
    except KeyError:    
        states[state] = [ df['happiness'][i] ] #if there is no existing entry set up a dictionary entry seed with the happiness. 

stateAvgs = {}  #take those values and find an average, again keyed by state name. 
for state in states.keys(): 
    stateAvgs[state] = numpy.mean(states[state])

statenames = pd.read_csv("stateAbrevToName.csv")  #We need to turn "VT" into "Vermont" to use the shapefile.     
stateFullName = {}
for i in range(len(statenames)): 
    try:
        stateFullName[statenames['US State:'][i]] = stateAvgs[statenames['Abbreviation:'][i]]
    except KeyError:
        pass
```

```python
# Plot averages on top of state shape files...

def plot_happiness():
    """Plot the basemap with state colors based on happiness index."""
    m = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=51, projection='lcc',lat_1=33,lat_2=45,lon_0=-95)
    shp_info = m.readshapefile('st99_d00/st99_d00','states',drawbounds=True)
    colors={}
    statenames=[]
    cmap = plt.cm.get_cmap('RdBu')

    vmin = min(df['happiness']); vmax = max(df['happiness']) # set range.

    for shapedict in m.states_info:
        statename = shapedict['NAME']
   
        if statename not in ['District of Columbia','Puerto Rico', 'Alaska' ,'Nevada','Hawaii']:
            happiness = stateFullName[statename]
        
            colors[statename] = cmap((happiness-vmin)/(vmax-vmin))[:3]
        statenames.append(statename)

    # cycle through state names, color each one.
    ax = plt.gca() # get current axes instance
    for nshape,seg in enumerate(m.states):
    
        if statenames[nshape] not in ['District of Columbia','Puerto Rico', 'Alaska' ,'Nevada','Hawaii']:
            color = rgb2hex(colors[statenames[nshape]]) 
            poly = Polygon(seg,facecolor=color,edgecolor=color)
            ax.add_patch(poly)

    plt.title('Happiness Index', fontsize = 25)
    plt.gcf().set_size_inches(23,25)

plot_happiness()

def plot_happiness_colorbar():
    cmap = plt.cm.get_cmap('RdBu')
    fig = pyplot.figure(figsize=(6,1)) # Define a figure and its axes
    ax = fig.add_axes([50, 80, 2, 0.2])
    norm = mpl.colors.Normalize(vmin = min(df['happiness']), vmax = max(df['happiness']))
    cb1 = mpl.colorbar.ColorbarBase(ax,cmap=cmap,norm=norm,orientation='horizontal')
    plt.title('Happiness Legend',fontsize = 20)
    plt.xlabel('Less Happy                                                            More Happy')

plot_happiness()
plot_happiness_colorbar()
```

```python
square_miles = []
for state in df_eagles['State']:
    index = 0
    for state2 in df_area['State']:
        if state2 == state:
            area = df_area['Area (Square Miles)'][index]
            square_miles.append(area)
        index = index + 1
    
df_eagles['Square_Miles'] = square_miles
```

```python
index = 0
smpe = []
for num in df_eagles['Eagle Population (2007)']:
    num_eagles = float(num)
    sm = square_miles[index]
    index = index + 1
    smpe.append(sm/num_eagles)
    
df_eagles['Miles/Eagles'] = smpe
```

##Notes: 
* There is a single state variable _x_, or energy reserves. It's time ranges from 0 to L, where L is the limit of energy reserves for the animal.  
* if _x_ reaches 0 the animal dies. 
* Foraging starts at time 0 and goes to time T, (dawn to dusk) 
* If the organism survives to time T with reserves _x_ then its reproductive value is $R(x)$. 
* There are two feeding options $u_1$ and $u_2$ 
* $z_1,z_2$ are the probability that the animal is killed by a predator between time $t$ and $t+1$. 
* If the animal is not killed then $p_i$ is the probability that it finds a food item. 
* Distributions of the food energy values are: 

|number | probability  |
|----------------------|
|1      | $\frac{1}{4}$|
|2      | $\frac{1}{2}$|
|3      | $\frac{1}{4}$|

* We let $p_1 < p_2$ and $z_1 < z_2$ 


```python
import numpy
import matplotlib.pyplot as plt

#Set up the probabilities associated with feeding and getting eaten

p_1 = 0.5; z_1 = 0.00
p_2 = 0.6; z_2 = 0.01

#Maximal reserve levels:
L = 20

#Create repository for calling prior results, V and H functions will try and call their inputs as keys. 
V_sols = {}
H_sols = {}

#Reward at dusk 

#R1 #Simplify this if the time is available to one single function. 
def R_1(x): 
    """The reward function. Takes the value of x at the end of the time interval and gives a boolian answer
    ex. R_1(11) = 1 
    """
    if x >= 10:
        res = 1
    else: 
        res = 0
    return res 

#R2, ignore the fact the name doesnt change
def R_2(x): 
    """The reward function. Takes the value of x at the end of the time interval and gives a boolian answer
    ex. R_1(11) = 1 
    """
    if x >= 1:
        res = 1
        return res
    if x == 0: 
        res = 0
        return res 
    
def R_3(x): 
    """The reward function. Takes the value of x at the end of the time interval and gives a boolian answer
    ex. R_1(11) = 1 
    """
    k = 2.1
    if x >= 0:
        res = k*x
    else: 
        res = 0
    return res 



#the x states: 
def X0(x):
    """X0 state
    """
    res = x - 1.00
    if res < 0.00:
        res = 0.00
    return res 

def X1(x):
    """X1 state
    """
    res = float(x) 
    return res 

def X2(x):
    """X2 state
    """
    res = x + 1.00
    if res > L:
        res = L 
    return res 

def X3(x):
    """X3 state
    """
    res = x + 2.00
    if res > L:
        res = L 
    return res 


#define the main functions, each option of i represents a different behavior choice. 
#The number of these will change based upon the factors. 

def H_Ui(x,t,i): 
    """Takes the given energy reserve(x) and a time(t) and a choice number(i) and returns the value of that decision. 
    ex. H_Ui(10,38,1) = 0.6719999999999999.
    """
    #V is the maximum value of H_Ui at the last time step t+1
    ID = "%f, %f, %f" %(x,t,i) 
    
    if ID in H_sols:
        return H_sols[ID]
    
    else: 
        if i == 1: 
            p = p_1; z = z_1
        else: 
            p = p_2; z = z_2
        
        H_sols[ID] = (1 - z) * ((1 - p)* V(X0(x),t + 1)[0] + \
                       (.25) *       p * V(X1(x),t + 1)[0] + \
                       (.5)  *       p * V(X2(x),t + 1)[0] + \
                       (.25) *       p * V(X3(x),t + 1)[0] )
        return H_sols[ID] 



def V(x, time): 
    """
    This is the optimality "finder." It takes the choices, runs through them and finds the most beneficial choice. 
    i.e H_Ui > all other choices. 
    ex. V(10,39) = 0.594. 
    """
    Total_time = 40
    
    ID = "%f, %f" %(x,time)
    
    if x == 0.0:
        V_sols[ID] = [x,1] 
        return V_sols[ID]
    
    if ID in V_sols:
        return V_sols[ID]
    
    
    if time == (Total_time - 1): #These are simply a slightly re-written version of the H_Ui functions, for t = T V = R_1
        u_1 = (1.00 - z_1) * ((1-p_1)  * R_1( X0(x) ) + \
                       (.25) * p_1     * R_1( X1(x) ) + \
                       (.5)  * p_1     * R_1( X2(x) ) + \
                       (.25) * p_1     * R_1( X3(x) ) )

        u_2 = (1.00 - z_2) * ((1-p_2)  * R_1( X0(x) ) + \
                       (.25) * p_2     * R_1( X1(x) ) + \
                       (.5)  * p_2     * R_1( X2(x) ) + \
                       (.25) * p_2     * R_1( X3(x) ) )
        if u_1 > u_2: 
            V_sols[ID] = [u_1,1]
            
        else: 
            V_sols[ID] = [u_2,2] 
            
        return V_sols[ID]
    
    else:     
        choices = [H_Ui(x,time,1), H_Ui(x,time,2)]
        
        selector = choices.index(max(choices)) 
        
        V_sols[ID] = [choices[selector], (selector + 1)]
        
        return V_sols[ID]
```

##Now we turn to optimal behavior strategy: 
###It's getting interesting now. 

* overall for loop goes by times from 0 to 40.
* inside of the time loop $p(x)$ scrolls through by $x$ value from 1 to L. 
* for each $x,t$ combination $V(x,t)$ is run to find the choice. 
* from that choice the set of $p$ and $z$ values are obtained. 
* those $p$ and $z$ values are then run throughthe series of P rules. 

```python
#x_0 = 8                                                           #This is the original cohort energy level. 

def Opt_Behav(x_0):
    P_sols = {}                                                       #Set up the P function dictionary
    Pred_sols = {}
                                                                  #seed the dictionary with the initial values. 
    for t in range(40+1):
        for x in range(L+1):
            x_t = "%i,%i"%(x,t) 
        
            if t == 0 and x == x_0:
                P_sols[x_t] = 1
            
            else:
                P_sols[x_t] = 0 

    for t in range(40 + 1):
        Pred_sols[t] = 0
       

    for t in range(40):
        for x in range(1 , L+1):
            if V(x,t)[1] == 1:
                p = p_1 
                z = z_1
            else:
                p = p_2 
                z = z_2 
        
            x_t      = "%i,%i"%(x,t)                                     #set up the key names for editing the P dictionary
            x0_tnext = "%i,%i"%(X0(x),t+1)
            x1_tnext = "%i,%i"%(X1(x),t+1)
            x2_tnext = "%i,%i"%(X2(x),t+1)
            x3_tnext = "%i,%i"%(X3(x),t+1)
                                                                     #P-Rules:
            P_sols[x0_tnext] = (1-z)*(1-p)*P_sols[x_t] + P_sols[x0_tnext]
            P_sols[x1_tnext] = .25*(1-z)*p*P_sols[x_t] + P_sols[x1_tnext]
            P_sols[x2_tnext] = .50*(1-z)*p*P_sols[x_t] + P_sols[x2_tnext]
            P_sols[x3_tnext] = .25*(1-z)*p*P_sols[x_t] + P_sols[x3_tnext]
            Pred_sols[t+1]   = z*P_sols[x_t] + Pred_sols[t+1]
            
    
            
    return [P_sols,Pred_sols]
```

```python
P_sols,Pres_sols = Opt_Behav(4)

def Alive(t):
    """Tells you the number of the original cohort alive at time t
    Ex. Alive(2) = .89 
    """
    alive = 0
    for x in range(L+1): 
        ID = "%i,%i"%(x,t) 
        alive += P_sols[ID]
    return alive

def p_u(t):
    """Tells you the percent of the original cohort that choice the uth choice. 
    Ex. p_u(4) = .76
    """
    p_u = 0
    for x in range(L+1):
        ID = "%i,%i"%(x,t)
        if V(x,t)[1] == 2:
            p_u += P_sols[ID]
    return p_u

def P_u(t): 
    res = p_u(t)/Alive(t)
    return res
```

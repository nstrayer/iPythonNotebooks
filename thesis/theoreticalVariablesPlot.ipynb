```python
import numpy
import matplotlib.pyplot as plt

X = numpy.linspace(0,2*numpy.pi, num = 1000)

Y_1 = numpy.sin(X - 2)/1.2

#Y_2 = numpy.sin(.5*X)
Y_2 = numpy.sin(.5*X) + numpy.sin((365/2)* X)/7


plt.plot(X,Y_1, '-r', linewidth = 1.5, label = 'Epilimnion temperature')

plt.plot(X,Y_2, '-g', linewidth = 1.5,label = 'Light Levels')


legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,fontsize=26)

#plt.legend([, , 'Mysis Condition', 'Resultant Pressure'], loc = 8)
plt.title('Theoretical Variable Interactions',fontsize=26)
plt.ylabel('Intensity (arbitrary scale)',fontsize=26)
plt.axis([0,2*numpy.pi,-1.2,2])
plt.gcf().set_size_inches(16,10)

frame = plt.gca()

frame.axes.get_xaxis().set_ticklabels(['Spring',' ','Summer',' ', 'Fall',' ', 'Winter'],fontsize=16)
frame.axes.get_yaxis().set_ticklabels(['','Low','','','','  ', 'High'],fontsize=16)

plt.savefig("variable_interactions",bbox_inches='tight', pad_inches=.2)
```

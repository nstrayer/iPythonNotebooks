___
##Linear Interpolation Function
First I have written a linear interpolation function that can be set to make $n$ dots along the line between two points. 

```python
def linear_interpolation(points):
    """ A nice simple linear interpolation function. 
    Takes input points [(a,b), (c,d)] and returns two lists of points interpolating those two lines: x_vec, y_vec
    """
    n = 100 #I chose to put this here as opposed to carrying it through the functions as a parameter. Eliminates some confusion. 
    x_points = [a[0] for a in points]
    y_points = [a[1] for a in points]
    f = interp1d(x_points,y_points)
    x_interp = numpy.linspace(x_points[0], x_points[-1], n) 
    y_interp = f(x_interp)

    return x_interp, y_interp
```

---
##Bezier Step function:
Preforms the interpolation of the lines for each "level" of the bezier curve.

```python
def bezier_step(x_vals, y_vals, t):
    """Completes one step of the bezier interpolation. Takes two vectors of the x and y values of the lines, then interpolates the lines between 
    them based upon the time. 
    Example usage: (newline_x, newline_y) = bezier_step([0,1,2,3,....],[3,2,4,....],40) 
    """
    test_X = []
    test_Y = []
    for i in range(0, (len(x_vals)/100 - 1)):
        X,Y = linear_interpolation([(x_vals[t + i*100],       y_vals[t + i*100]),\
                                (x_vals[(t+100) + i*100], y_vals[(t+100) + i*100])])
        test_X.extend(X)
        test_Y.extend(Y)
    return test_X, test_Y
    #plot(test_X,test_Y)
```

---
##Sketch the whole bezier curve at given time function:
1. create two empty x and y vectors
2. start if loop
3. run the bezier function on the base lines, append the result to the x and y list. 
4. restart x,y and repeate bezier. 


```python
def bezier_curve_sketch(x_vals,y_vals,time):
    """Takes the initialy paratarized lines between points and plots all of the following paramaterized lines
    sample usage: bezier_curve_sketch(X coordinates, Y coordinates, time): gives you a graph of the bezier construction
    at that moment. 
    """
    x_vec = []
    y_vec = []
    for _ in range(0,(len(x_vals)/100 - 1)):
        X,Y = bezier_step(x_vals,y_vals,time) 
        x_vec.append(X)
        y_vec.append(Y) 
        x_vals = X #Reseed the bezier step function with these new values. 
        y_vals = Y
    return x_vec, y_vec
```
